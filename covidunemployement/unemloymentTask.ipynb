```python
# Convert to Datetime format   
df['Date'] = pd.to_datetime(df['Date'])
df_2020['Date'] = pd.to_datetime(df_2020['Date'])
```

```python
df.info()
df_2020.info()
```

```python
#adding new colums that ca be easier for visuallizing
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df_2020['Year'] = df_2020['Date'].dt.year
df_2020['Month'] = df_2020['Date'].dt.month_name()
```

```python
#EDA
df.describe()
```

```python
df_2020.describe()
```

```python
#avarage unemployement rate of each date
trend=df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
trend
```

```python
#line chart 
plt.plot(trend.index,trend.values,marker='o',color='red')
plt.xlabel('year-month')
plt.ylabel("Average Unemployment rate(%)")
plt.title("Unemployment over time",color="blue",)
plt.grid()
```

- The unemployment rate remained relatively stable during 2019.
- A sharp increase is observed around March–April 2020, corresponding to the COVID-19 lockdown period.
- After the peak, unemployment gradually declined but remained higher than pre-pandemic levels.

```python
trend_2020=df_2020.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
```

```python
plt.figure(figsize=(10,8))
plt.barh(trend_2020.index,trend_2020.values,color='darkblue')
plt.title('Average Unemployment Rate by State/Region in 2020 ',fontsize=20,color='red')
plt.ylabel("states",fontsize=16)
plt.xlabel("rate of unemployment",fontsize=16)
plt.grid(axis='x')
```

- Some regions recorded consistently higher unemployment rates than others.
- The variation suggests that unemployment was affected differently across regions.
- States with higher unemployment may require targeted employment policies.

```python
count=df.groupby('Area')['Estimated Unemployment Rate (%)'].mean()
```

```python
plt.figure(figsize=(4,4))
plt.bar(count.index,count.values,width=0.3,color='green')
plt.xlabel("Area")
plt.ylabel('rate of unemplyment(%)')
plt.title("Rural vs Urban",color="red",fontsize=16)
```

The average unemployment rate differs between rural and urban areas.
Urban employment was more affected during the COVID-19 period due to disruptions in industries and services.
The comparison indicates that the impact of unemployment varied by area.

```python
monthlytrend_2020=df_2020.groupby('Month')['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(10,6))
plt.plot(monthlytrend_2020.index,monthlytrend_2020.values,marker='o')
plt.xlabel('Months',fontsize=16)
plt.ylabel('Average unemployment rate(%)',fontsize=16)
plt.title("Avarage unemployment rates by months in 2020",color='red',fontsize=20)
plt.grid()
```

- The unemployment rate increased sharply during April and May 2020, coinciding with the nationwide COVID-19 lockdown.
- After reaching its peak, the unemployment rate gradually declined in the following months as economic activities resumed.
- The graph clearly shows the significant impact of the COVID-19 pandemic on employment during 2020.

# Conclusion

- The COVID-19 pandemic had a significant impact on unemployment rates.
- Unemployment peaked during the lockdown period and gradually recovered afterward.
- Regional differences indicate that economic recovery was uneven.
- The analysis can help policymakers identify regions that need employment support.
